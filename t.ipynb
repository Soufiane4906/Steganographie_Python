```python
from tkinter import *
from tkinter import filedialog, messagebox
from PIL import Image, ImageTk
import os
from stegano import lsb

# Créer la fenêtre principale
root = Tk()
root.title("Steganographie - Mini Projet de Traitement d'image")
root.geometry("750x600+100+50")
root.resizable(False, False)
root.configure(bg="#1e1e2f")

photo = None  # Variable globale pour l'image sélectionnée
filename = None
secret = None

# Fonction pour afficher une image
def Aff_Image():
    global filename, photo
    filename = filedialog.askopenfilename(
        initialdir=os.getcwd(),
        title="Sélectionner une image",
        filetypes=[("Fichiers PNG", "*.png"), ("Fichiers JPG", "*.jpg"), ("Tous les fichiers", "*.*")]
    )
    if filename:
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            IMG = Image.open(filename)
            IMG = IMG.resize((250, 250))  # Redimensionner l'image pour l'affichage
            photo = ImageTk.PhotoImage(IMG)
            lbl.configure(image=photo)
            lbl.image = photo
        else:
            messagebox.showerror("Erreur", "Veuillez sélectionner une image au format PNG ou JPG.")

# Fonction pour cacher un message dans l'image
def cacher_Image():
    global secret
    message = text1.get(1.0, END).strip()
    if not filename:
        messagebox.showwarning("Erreur", "Veuillez sélectionner une image avant de cacher un message.")
        return
    if not message:
        messagebox.showwarning("Erreur", "Veuillez entrer un message à cacher.")
        return
    try:
        secret = lsb.hide(filename, message)
        messagebox.showinfo("Succès", "Le message a été caché avec succès dans l'image.")
    except Exception as e:
        messagebox.showerror("Erreur", f"Une erreur s'est produite : {e}")

# Fonction pour sauvegarder l'image avec le message caché
def sauv():
    global secret
    if not secret:
        messagebox.showwarning("Erreur", "Aucune image avec un message caché à sauvegarder.")
        return
    save_path = filedialog.asksaveasfilename(
        defaultextension=".png",
        filetypes=[("Fichiers PNG", "*.png")],
        title="Enregistrer sous"
    )
    if save_path:
        try:
            secret.save(save_path)
            messagebox.showinfo("Succès", "Image sauvegardée avec succès.")
        except Exception as e:
            messagebox.showerror("Erreur", f"Une erreur s'est produite lors de la sauvegarde : {e}")

# Fonction pour révéler un message caché dans une image
def Aff_Img():
    if not filename:
        messagebox.showwarning("Erreur", "Veuillez sélectionner une image pour révéler un message.")
        return
    try:
        message_clair = lsb.reveal(filename)
        if message_clair:
            text1.delete(1.0, END)
            text1.insert(END, message_clair)
        else:
            messagebox.showinfo("Information", "Aucun message caché trouvé dans l'image.")
    except Exception as e:
        messagebox.showerror("Erreur", f"Une erreur s'est produite : {e}")

# Créer une interface utilisateur soignée
f = Frame(root, bg="black", width=900, height=900, relief=GROOVE)
f.place(x=20, y=80)
lbl = Label(f, bg="black")
lbl.place(x=25, y=25)

f2 = Frame(root, bg="white", width=380, height=300, relief=GROOVE)
f2.place(x=400, y=80)
text1 = Text(f2, font="Arial 14", bg="white", fg="black", relief=GROOVE, wrap=WORD)
text1.place(x=0, y=0, width=360, height=280)
scrollbar1 = Scrollbar(f2)
scrollbar1.place(x=360, y=0, height=280)
scrollbar1.configure(command=text1.yview)
text1.configure(yscrollcommand=scrollbar1.set)

f3 = Frame(root, bg="#1e1e2f", width=750, height=80, relief=GROOVE)
f3.place(x=0, y=400)
Button(f3, text="Ouvrir", width=12, font="Arial 12 bold", bg="#3498db", fg="white", command=Aff_Image).place(x=50, y=20)
Button(f3, text="Cacher", width=12, font="Arial 12 bold", bg="#2ecc71", fg="white", command=cacher_Image).place(x=200, y=20)
Button(f3, text="Afficher", width=12, font="Arial 12 bold", bg="#e74c3c", fg="white", command=Aff_Img).place(x=350, y=20)
Button(f3, text="Sauvegarder", width=12, font="Arial 12 bold", bg="#f39c12", fg="white", command=sauv).place(x=500, y=20)

# Ajouter des instructions
instructions = Label(root, text="Instructions :\n1. Cliquez sur 'Ouvrir' pour choisir une image.\n2. Entrez un message dans la zone de texte.\n3. Cliquez sur 'Cacher' pour insérer le message dans l'image.\n4. Cliquez sur 'Sauvegarder' pour enregistrer l'image modifiée.\n5. Cliquez sur 'Afficher' pour révéler un message caché.",
                     bg="#1e1e2f", fg="white", font="Arial 10", justify=LEFT, anchor="w")
instructions.place(x=20, y=500, width=710, height=120)

# Lancer l'application
root.mainloop()
```
